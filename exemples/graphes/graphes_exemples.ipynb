# Exemples de graphes typiques

```python
# imports canoniques
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
```

## Trigonométrie

### Cosinus et sa dérivée

```python
p = 2  # nombre de périodes voulues
t = np.linspace(
    -p * pi, p * pi, p * 60
)  # échantillonage de valeurs pour theta entre -nπ et nπ, avec suffisamment de points

fig, ax = plt.subplots(
    figsize=(12, 6)
)  # création de la Figure et de l'objet Axes, avec sa taille en pouces

ax.plot(t, np.cos(t), label="$cos(t)$")  # on trace
ax.plot(t, -np.sin(t), label="$-sin(t)$")  # on trace la dérivée

# réduire les ticks sur les ordonnées
ax.set_yticks((-1, 0, 1))

# mettre les ticks en x en multiples de pi
ax.set_xticks([pi * k for k in range(-p, p + 1)])  # valeurs où mettre les ticks
ax.set_xticklabels(
    [f"{k}π" for k in range(-p, p + 1)]
)  # texte à substituer aux valeurs numériques

ax.legend();
```

## Pendule simple

`x` est la distance parcourue par le pendule entre la verticale et l’angle θ

- acceleration = $x'' = -g \times sin(θ)$
- $x = L\times θ$
- donc $L \times θ'' = -g \times sin(θ)$
- donc $θ'' = (-g/L) \times sin(θ)$


```python
g = 9.8
L = 1

ω = np.sqrt(g / L)
theta_0 = 0.31


def theta(t):
    return theta_0 * np.cos(ω * t)
```

```python
theta(0)  # simple test de la fonction
```

```python
# représentation graphique simple et rapide
t = np.linspace(
    0, 10, 1000
)  # tableau de valeurs de t, entre 0 et 10, avec 1000 valeurs
y = theta(t)  # y est un tableau contenant theta(t) pour chaque élément du tableau t.

plt.plot(t, y)
plt.show()
```

```python
# représentation graphique un peu plus élaborée
# t et y = θ(t) sont déjà définis plus haut

fig, ax = plt.subplots(figsize=(16, 8))  # création de la Figure et de l'Axes

ax.plot(t, y, label="θ(t)")  # tracé de la fonction
ax.plot(t, [theta_0] * len(t), "--", c="orange", label="θ₀")  # tracé du max
ax.plot(t, [-theta_0] * len(t), "--", c="orange", label="—θ₀")  # tracé du min

ax.legend();
```

## Ballistique : parabole de sécurité

```python
x = np.linspace(0, 0.1, 100)
v0 = 1
g = 9.81

# Trajectoires :
alpha = 80 * (pi / 180)
y1 = -g * x ** 2 / (2 * v0 ** 2 * np.cos(alpha) ** 2) + v0 * np.tan(alpha) * x

alpha = 70 * (pi / 180)
y2 = -g * x ** 2 / (2 * v0 ** 2 * np.cos(alpha) ** 2) + v0 * np.tan(alpha) * x

alpha = 60 * (pi / 180)
y3 = -g * x ** 2 / (2 * v0 ** 2 * np.cos(alpha) ** 2) + v0 * np.tan(alpha) * x

alpha = 50 * (pi / 180)
y4 = -g * x ** 2 / (2 * v0 ** 2 * np.cos(alpha) ** 2) + v0 * np.tan(alpha) * x

alpha = 40 * (pi / 180)
y5 = -g * x ** 2 / (2 * v0 ** 2 * np.cos(alpha) ** 2) + v0 * np.tan(alpha) * x

alpha = 30 * (pi / 180)
y6 = -g * x ** 2 / (2 * v0 ** 2 * np.cos(alpha) ** 2) + v0 * np.tan(alpha) * x

# Parabole de sécurité :
ys = -(g * x ** 2) / 2 * (v0 ** 2) + v0 ** 2 / (2 * g)

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x, y1, x, y2, x, y3, x, y4, x, y5, x, y6, x, ys)
ax.set_xlim(0, 0.1)
ax.set_ylim(0, 0.06);
```

## Energie potentielle d'un balancier de type métronome

```python
import numpy as np
import matplotlib.pyplot as plt

theta=np.linspace(-(np.pi)/2, (np.pi)/2, 100)
C=0.79e-2
P=10e-3*10
l=0.1
Ep=(1/2)*C*theta**2+P*l*np.cos(theta)

fig, ax = plt.subplots(figsize=(12,8))

ax.plot(theta,Ep, label='Energie potentielle')
ax.legend();
```
