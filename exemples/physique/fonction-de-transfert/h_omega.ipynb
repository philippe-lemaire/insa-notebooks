```python
def H(omega, omega_0=1):
    z = complex(1, omega/omega_0)
    return 1/z
```

```python
def mod_H(omega, omega_0=1):
    val = np.sqrt(1 + (omega/omega_0)**2)
    return 1/val
    
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 30))
omega = np.linspace(0, 10**3, 10**6)

y = mod_H(omega)

(ax1, ax2, ax3) = axs

ax1.plot(omega,y )

ax2.set_xscale('log')
ax2.plot(omega, y)

ax3.set_xscale('log')
ax3.plot(omega, 20*np.log(y))


for ax in axs:
    ax.set_title('title')
    ax.set_xlabel('ω')
    ax.set_ylabel('mod_H')
    

ax3.set_ylabel('G')


```
