```python
import matplotlib.pyplot as plt

from IPython.display import clear_output, display

import numpy as np
```

```python
t = np.arange(0.0, 5.0, 0.01)

fig, ax = plt.subplots(figsize=(10, 6))

for i in range(1000):
    #print(f"Itération numéro {i}.")
    ax.cla()  # effacer l'ecran
    ax.set_xlim(0, 5)
    ax.set_ylim(-1.3, 1.3)
    ax.plot(t, np.cos(3 * t + i / 10))
    display(fig)
    #clear_output(wait=True)  # permet d'afficher dans le même graphique
```

```python
# Create figure and subplot

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1) 

# Define and update plot

for i in range(100):
    ax.set_xlim(0, 50)
    ax.plot(i, 2, marker='o',c='r')
    display(fig)    
    clear_output(wait = True)
    plt.pause(0.2)
```
