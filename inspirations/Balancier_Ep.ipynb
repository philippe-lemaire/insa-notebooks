```python
import numpy as np
import math
from math import pi
```

```python
C = 0.0079 # constante de torsion (Kg.m².s-2)
m = 0.01 # masse (Kg)
g = 10 # gravité (m/s²)
l = 0.1 # longueur (m)
```

```python
steps = 200       # nombre de pas pour les angles 
theta = np.linspace(-pi/2, pi/2, steps) # tableau des angles
```

```python
Ep = 1/2 * C * theta**2 + m * g * l * (np.cos(theta) - 1)
```

```python
import matplotlib.pyplot as plt
```

```python
fig , ax = plt.subplots(figsize=(15 , 6))
ax.plot(theta, Ep , color = 'green' , label='Energie potentielle')
ax.grid(True)
ax.set_xlabel('theta (rd)')
ax.set_ylabel('Ep (J)')
ax.set_title("Balancier \n", size = 'xx-large')
ax.grid(True)
ax.legend()
ax.set_xticks([-pi/2, -pi/4 , 0 , pi/4 , pi/2] , ['$-\pi /2$' , '$-\pi /4$' , 0 , '$-\pi /4$' , '$\pi /2$'])

ax_top = ax.secondary_xaxis('top', functions=(np.rad2deg, np.deg2rad))
ax_top.set_xlabel('Angle [°]');
ax_top.set_xticks([-90 , -45 , 0 , 45, 90])
fig.savefig('Ep_balancier.png')
```
